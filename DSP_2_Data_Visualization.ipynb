## Columns Renaming

```python
df.columns = [col.replace(' ', '_').lower() for col in df.columns]
df.head()
```

## To Datetime datatype

```python
pd.to_datetime(df["order_date"])
pd.to_datetime(df["ship_date"])
```

```python
df[["order_date", "ship_date"]].head().info()
```

```python
df = df.sort_values(
    by=["order_id", "order_date", "ship_date"],
    ascending=[True, True, True],
    ignore_index=True
)

df.head(20)
```

## Remove "$" in ["sales", "profit", "shipping_cost" ]

```python
dollar_cols = ["sales", "profit", "shipping_cost"]
for col in dollar_cols:
  df[col] = (
      df[col].str.replace("[^\d.]", "", regex = True)
  )
  df.rename(columns={col: col+"($)"}, inplace=True)
print("Formatting & Renaming, Done!")
```

## Monthly Sales Trends

```python
monthly_sum_sales = df.groupby("months").agg({"sales($)":"sum"})
monthly_sum_sales
```

```python
ordered_months = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

Month_CI = pd.CategoricalIndex(
    monthly_sum_sales.index,
    categories=ordered_months,
    ordered=True
)

Month_CI
```

```python
monthly_sum_sales.index = Month_CI
monthly_sum_sales = monthly_sum_sales.sort_index()
monthly_sum_sales
```

```python
import matplotlib.ticker as ticker
```

```python
# Create line chart for monthly sales
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(
    monthly_sum_sales.index,
    monthly_sum_sales["sales($)"],
    marker='*',
    mfc='red', # marker face color
    mec='red', # marker edge color
    linestyle='-'
)

ax.set_title('Monthly Sales Trends')
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.grid(True)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '$' + '{:,.2f}'.format(x/1000) + 'K'))

# Show the line chart
fig.show()
```

## 1.&nbsp; **Profit vs Shipping Cost: Scatter plot**

```python
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=df, x='shipping_cost($)', y='sales($)', hue='region')
plt.title('Shipping Cost vs Sales with Region')
plt.xlabel('Shipping Cost ($)')
plt.ylabel('Sales ($)')
plt.legend(title='Region', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

## 4.&nbsp; **Monthly Trends: Bar chart for quantity per month**

```python
# Monthly Trends: Plot line charts or bar charts to visualize how profit, or quantity changes month by month.

months = df['months']
quantity = df['quantity']

sum_quantity = quantity.groupby(months).sum().sort_values()

plt.bar(
    sum_quantity.index,
    sum_quantity.values

)

plt.title('Monthly Trends in Quantity')
plt.xlabel('Month')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```


Kesimpulan dan Insight yang bisa didapat dari visualisasi daiatas adalah

* Visualisasi bar chart ini menunjukkan total kuantitas produk yang terjual setiap bulan. Dengan mengamati tinggi rendahnya batang pada grafik, kita dapat melihat fluktuasi penjualan sepanjang tahun.

* Dengan melihat tinggi batang, kita dapat dengan mudah mengidentifikasi bulan dengan penjualan tertinggi dan terendah. Ini membantu dalam merencanakan strategi pemasaran dan persediaan untuk bulan-bulan mendatang.

* Dengan memahami tren bulanan dalam kuantitas, perusahaan dapat mengambil keputusan yang lebih baik dalam hal pengelolaan inventaris, strategi pemasaran, dan perencanaan keuangan untuk meningkatkan kinerja penjualan keseluruhan.


## 5.&nbsp; **Region-wise Analysis: Bar charts**

```python
# Region-wise Analysis: Create bar charts or stacked bar charts to compare sales, profit, or quantity across different regions.

region_data = df['region']
sales_data = df['sales($)']

sum_sales = sales_data.groupby(region_data).sum().sort_values()

plt.bar(
    sum_sales.index,
    sum_sales.values

)

plt.title('Sales Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```


Kesimpulan dan Insight yang bisa didapat dari visualisasi daiatas adalah

* Visualisasi bar chart ini menunjukkan total penjualan berdasarkan wilayah. Setiap batang mewakili total penjualan di masing-masing wilayah, memberikan gambaran yang jelas tentang kinerja penjualan di setiap area.

* Jika satu atau dua wilayah menunjukkan penjualan yang jauh lebih tinggi dibandingkan wilayah lainnya, ini menunjukkan bahwa wilayah tersebut mungkin merupakan pasar utama bagi perusahaan. Misalnya, jika wilayah "Central" memiliki penjualan yang sangat tinggi, perusahaan mungkin perlu fokus pada pengembangan lebih lanjut di wilayah tersebut.

* Dengan memahami distribusi penjualan berdasarkan wilayah, perusahaan dapat membuat keputusan yang lebih baik dalam hal pengelolaan sumber daya, strategi pemasaran, dan pengembangan produk untuk meningkatkan kinerja penjualan secara keseluruhan.
